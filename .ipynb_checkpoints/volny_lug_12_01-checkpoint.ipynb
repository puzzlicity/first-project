1. –°–æ–∑–¥–∞–µ–º –º–æ–¥–µ–ª—å –ª–∏–Ω–µ–π–Ω–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏. –î–ª—è –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏—è –º–æ–¥–µ–ª–∏ –æ—Ç–±–∏—Ä–∞–µ–º –≤ –∫–∞—á–µ—Å—Ç–≤–µ —Ü–µ–ª–µ–≤–æ–≥–æ –ø—Ä–∏–∑–Ω–∞–∫–∞:
* —É–¥–æ–π,
  
–∞ –≤ –∫–∞—á–µ—Å—Ç–≤–µ –≤—Ö–æ–¥–Ω—ã—Ö -

* –≠–ö–ï,
* —Å—ã—Ä–æ–π –ø—Ä–æ—Ç–µ–∏–Ω,
* –ø–æ—Ä–æ–¥—É,
* —Ç–∏–ø –ø–∞—Å—Ç–±–∏—â–∞,
* –ø–æ—Ä–æ–¥—É –ø–∞–ø—ã.

  –°–ü–û –Ω–µ –±–µ—Ä–µ–º, –ø–æ—Ç–æ–º—É —á—Ç–æ —ç—Ç–æ—Ç –ø—Ä–∏–∑–Ω–∞–∫ –æ–∫–∞–∑–∞–ª—Å—è –≤–∑–∞–∏–º–æ–∑–∞–≤–∏—Å–∏–º –æ—Ç –¥—Ä—É–≥–∏—Ö (–≤–æ–∑—Ä–∞—Å—Ç–∞, –≠–ö–ï).
  –ñ–∏—Ä–Ω–æ—Å—Ç—å –∏ –±–µ–ª–æ–∫ –Ω–µ –±–µ—Ä–µ–º, –ø–æ—Ç–æ–º—É —á—Ç–æ –æ–Ω–∏ –ø–æ—á—Ç–∏ –≤—Å–µ–≥–¥–∞ –æ–¥–∏–Ω–∞–∫–æ–≤—ã–µ.
  –í–∫—É—Å –Ω–µ –±–µ—Ä–µ–º, –ø–æ—Ç–æ–º—É —á—Ç–æ —ç—Ç–æ —Ç–æ–∂–µ —Ü–µ–ª–µ–≤–æ–π –ø—Ä–∏–∑–Ω–∞–∫ –≤ –¥–∞–ª—å–Ω–µ–π—à–µ–º.

```python
df_m = df1[['udoy, kg', 'eke', 'syroy_protein', 'vozrast', 'poroda', 'pastbische', 'poroda_papy']]
X = df_m.drop('udoy, kg' , axis=1)
y = df_m['udoy, kg']

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    random_state = RANDOM_STATE)

cat_col_names = ['vozrast', 'poroda', 'pastbische', 'poroda_papy']

num_col_names = ['eke', 'syroy_protein']


encoder = OneHotEncoder(sparse_output=False, drop='first')

X_train_ohe = encoder.fit_transform(X_train[cat_col_names])

X_test_ohe = encoder.transform(X_test[cat_col_names])

encoder_col_names = encoder.get_feature_names_out()


scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train[num_col_names])

X_test_scaled = scaler.transform(X_test[num_col_names])


X_train_ohe = pd.DataFrame(X_train_ohe, columns=encoder_col_names)

X_test_ohe = pd.DataFrame(X_test_ohe, columns=encoder_col_names)


X_train_scaled = pd.DataFrame(X_train_scaled, columns = num_col_names)

X_test_scaled = pd.DataFrame(X_test_scaled, columns = num_col_names)

X_train = pd.concat([X_train_ohe, X_train_scaled], axis=1)

X_test = pd.concat([X_test_ohe, X_test_scaled], axis=1)


model_lr = LinearRegression()
model_lr.fit(X_train, y_train)

predictions = model_lr.predict(X_test)

residuals = y_test - predictions


fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10,5))
axes[0].hist(residuals)
axes[0].set_title('–ì–∏—Å—Ç–æ–≥—Ä–∞–º–º–∞ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –æ—Å—Ç–∞—Ç–∫–æ–≤')
axes[0].set_xlabel('–û—Å—Ç–∞—Ç–∫–∏')
axes[1].scatter(predictions, residuals)
axes[1].set_xlabel('–ü—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è –º–æ–¥–µ–ª–∏')
axes[1].set_ylabel('–û—Å—Ç–∞—Ç–∫–∏')
axes[1].set_title('–ê–Ω–∞–ª–∏–∑ –¥–∏—Å–ø–µ—Ä—Å–∏–∏')
plt.show()

mse = mean_squared_error(predictions, y_test)
print(mse)
rmse  = np.sqrt(mean_squared_error(predictions, y_test))
print(rmse)
mae = mean_absolute_error(y_test, predictions)
mae
```

```python
r2 = r2_score(y_test, predictions)
r2
```

–ü—Ä–æ–±–ª–µ–º —Å –≤—ã–±—Ä–æ—Å–∞–º–∏ –Ω–µ—Ç. –ë–∏–º–æ–¥–∞–ª—å–Ω–æ—Å—Ç–∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è —Ç–æ–∂–µ –Ω–µ—Ç.

–≠—Ç–æ –∑–Ω–∞—á–∏—Ç, —á—Ç–æ –¥–ª—è —É–ª—É—á—à–µ–Ω–∏—è —Ä–∞–±–æ—Ç—ã –º–æ–¥–µ–ª–∏ —Å—Ç–æ–∏—Ç —Ç—Ä–∞–Ω—Å—Ñ–æ—Ä–º–∏—Ä–æ–≤–∞—Ç—å –µ–µ –ø—Ä–∏–∑–Ω–∞–∫–∏ (–ø–µ—Ä–µ–≤–µ—Å—Ç–∏ —á–∏—Å–ª–æ–≤–æ–π –ø—Ä–∏–∑–Ω–∞–∫ –≤ –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã–π, –Ω–∞–ø—Ä–∏–º–µ—Ä).

R2 —Ä–∞–≤–Ω–æ 0.74  - —ç—Ç–æ –∑–Ω–∞—á–∏—Ç, —á—Ç–æ –≤ 74 –ø—Ä–æ—Ü–µ–Ω—Ç–∞—Ö –º–æ–¥–µ–ª—å –ø—Ä–µ–¥—Å–∫–∞–∑—ã–≤–∞–µ—Ç —Ä–µ–∑—É–ª—å—Ç–∞—Ç –ª—É—á—à–µ —Å—Ä–µ–¥–Ω–µ–≥–æ –∑–Ω–∞—á–µ–Ω–∏—è, –æ–¥–Ω–∞–∫–æ –≤ 26% –ø—Ä–æ—Ü–µ–Ω—Ç–∞—Ö —Å–ª—É—á–∞—è—Ö –æ–Ω–∞ –æ—à–∏–±–∞–µ—Ç—Å—è —Å–∏–ª—å–Ω–µ–µ, —á–µ–º —Å—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ —Ü–µ–ª–µ–≤–æ–≥–æ –ø—Ä–∏–∑–Ω–∞–∫–∞.

2. –°–æ–∑–¥–∞–µ–º –Ω–æ–≤—É—é –º–æ–¥–µ–ª—å –ª–∏–Ω–µ–π–Ω–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏, –≤–∫–ª—é—á–∞–µ–º –≤ –Ω–µ–µ –°–ü–û. –î–ª—è —Ç–æ–≥–æ, —á—Ç–æ–±—ã –∏—Å–∫–ª—é—á–∏—Ç—å –Ω–µ–ª–∏–Ω–µ–π–Ω–æ—Å—Ç—å —Å–≤—è–∑–∏ –°–ü–û –∏ —É–¥–æ—è, –ø–µ—Ä–µ–≤–æ–¥–∏–º –°–ü–û –≤ –±–∏–Ω–∞—Ä–Ω—ã–π –ø—Ä–∏–∑–Ω–∞–∫. –ü–æ –¥–∏–∞–≥—Ä–∞–º–º–µ —Ä–∞—Å—Å–µ—è–Ω–∏—è –º—ã –≤–∏–¥–∏–º, —á—Ç–æ –ø—Ä–∏–Ω—Ü–∏–ø–∏–∞–ª—å–Ω—ã–µ —Ä–∞–∑–ª–∏—á–∏—è –ª–µ–∂–∞—Ç—å –ø–æ –≥—Ä–∞–Ω–∏—Ü–µ –∑–Ω–∞—á–µ–Ω–∏—è 0.9.

```python
def cluster_func(row):
    spo = row['spo']
    if spo>=0.9:
        return 2
    else:
        return 1
    
df1['cluster'] = df1.apply(cluster_func, axis=1)
```

```python
df_ml = df1[['eke_2', 'syroy_protein', 'pastbische',  'vkus', 'poroda_papy', 'vozrast',]]

X = df_ml.drop('vkus' , axis=1)
y = df_ml['vkus']

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    random_state = RANDOM_STATE)


cat_col_names = [ 'vozrast', 'pastbische', 'poroda_papy']

num_col_names = ['eke_2', 'syroy_protein']


encoder = OneHotEncoder(sparse_output=False, drop='first')

X_train_ohe = encoder.fit_transform(X_train[cat_col_names])

X_test_ohe = encoder.transform(X_test[cat_col_names])

encoder_col_names = encoder.get_feature_names_out()


scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train[num_col_names])

X_test_scaled = scaler.transform(X_test[num_col_names])


X_train_ohe = pd.DataFrame(X_train_ohe, columns=encoder_col_names)

X_test_ohe = pd.DataFrame(X_test_ohe, columns=encoder_col_names)


X_train_scaled = pd.DataFrame(X_train_scaled, columns = num_col_names)

X_test_scaled = pd.DataFrame(X_test_scaled, columns = num_col_names)

X_train = pd.concat([X_train_ohe, X_train_scaled], axis=1)

X_test = pd.concat([X_test_ohe, X_test_scaled], axis=1)

clf = LogisticRegression()
clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)

```

```python
acc = accuracy_score(y_test, y_pred)
acc
```
